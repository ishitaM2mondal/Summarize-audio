```python
from sentence_transformers import SentenceTransformer
```

```python
import pandas as pd
```

```python
df = pd.read_excel('/content/Research_Project_Dataset_2.xlsx')
```

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.metrics import silhouette_score
```

```python
# model = SentenceTransformer('all-MiniLM-L6-v2')  # A lightweight BERT-based model
#model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
model = SentenceTransformer('all-mpnet-base-v2')
embeddings = model.encode(df['Summary'].tolist(), show_progress_bar=True)

# Step 3: Apply Clustering (e.g., K-Means)
num_clusters = 2
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
df['cluster'] = kmeans.fit_predict(embeddings)

# Step 4: Evaluate Clustering
silhouette_avg = silhouette_score(embeddings, df['cluster'])
print(f"Silhouette Score: {silhouette_avg}")
```

```python
from sklearn.cluster import AgglomerativeClustering

# Assuming you have your embeddings and df DataFrame

# Step 3: Apply Agglomerative Clustering
num_clusters = 2
agglomerative_clustering = AgglomerativeClustering(n_clusters=num_clusters)
df['cluster'] = agglomerative_clustering.fit_predict(embeddings)

# Step 4: Evaluate Clustering
silhouette_avg = silhouette_score(embeddings, df['cluster'])
print(f"Silhouette Score: {silhouette_avg}")
```
